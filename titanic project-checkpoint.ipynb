```python
def male_female_child(passenger):
    age,sex = passenger
    
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex',]].apply(male_female_child,axis=1)
```

```python
titanic_df[0:10]
```

```python
sns.factorplot(x='Pclass', data=titanic_df, hue='person', order=[1,2,3], hue_order=['child', 'female', 'male'], kind='count')
```

```python
#what deck were the passengers on and how does that that relate to their class?
```

```python
titanic_df.head()
```

```python
deck = titanic_df['Cabin'].dropna()
deck.head()
```

```python
levels = []

for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)

cabin_df.columns = ['Cabin']

sns.factorplot(x='Cabin', data=cabin_df, kind='count', palette='winter_d')
    
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']
sns.factorplot(x='Cabin', data=cabin_df, kind='count', palette='summer_d')
```

```python
#who was alone and who was with family?
```

```python
titanic_df.head()
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
titanic_df['Alone']
```

```python
titanic_df['Alone'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df.loc[titanic_df['Alone'] > 0, 'Alone'] = 1
titanic_df.loc[titanic_df['Alone'] == 0, 'Alone'] = 0
titanic_df['Alone'] = titanic_df['Alone'].replace({0: 'Alone', 1: 'With Family'})
```

```python
titanic_df.head()
```

```python
sns.factorplot(x='Alone', data=titanic_df, kind='count', palette='winter_d')
```
